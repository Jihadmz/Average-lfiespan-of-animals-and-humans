```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.futurelearn.com/info/courses/maths-linear-quadratic/0/steps/12167')
```

```python
page.close()
```

```python
print(page.content)
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
print(soup.table)
print(len(soup.table))
table = soup.table
```

```python
print(table.td)
tds = table.findAll('td')
print(len(tds))

td_names=[]

for x in range(0,130,2):
    tds[x] = tds[x].text
    td_names.append(tds[x])
    
print(len(td_names))   
print(td_names)
```

```python
td_age = []
for x in range(1,130,2):
    tds[x] = tds[x].text
    td_age.append(tds[x])
    
print(td_age)    
```

```python
content = pd.DataFrame({'Animal':td_names,'Average Life Span of Animal (in years)':td_age})
```
